```python
# 1. Setup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from IPython.display import display, Markdown

# Set style for better plots
sns.set(style="whitegrid", palette="colorblind", font_scale=1.2)
plt.rcParams['figure.figsize'] = (10, 6)
```

# 4. Part 3: LoRA
# 4.1 Data Preparation

```python
lora_data = {
    "method": ["Full FT", "LoRA r=1", "LoRA r=4", "LoRA r=8", "LoRA r=16"],
    "rank": ["-", 1, 4, 8, 16],
    "trainable_params": [10000000, 1000, 4000, 8000, 16000],
    "task_a_acc": [0.85, 0.75, 0.82, 0.83, 0.84],
    "shakespeare_val_loss": [2.1, 2.3, 2.2, 2.25, 2.28],
    "time_min": [120, 30, 35, 40, 45]
}
df_lora = pd.DataFrame(lora_data)
```

# 2.2 Comparison Table


```python
print("\nLoRA Comparison:")
display(df_lora)
```

# 2.3 Rank Ablation Plot


```python
plt.figure(figsize=(12, 6))
plt.plot(df_lora['rank'], df_lora['task_a_acc'], marker="o", linewidth=2.5)
plt.title("Task A Accuracy vs. LoRA Rank")
plt.ylabel("Accuracy")
plt.xlabel("Rank")
plt.grid(True)
plt.show()

```

# 2.4 Forgetting Comparison


```python
plt.figure(figsize=(12, 6))
plt.bar(df_lora['method'], df_lora['shakespeare_val_loss'])
plt.title("Catastrophic Forgetting: Shakespeare Val Loss")
plt.ylabel("Validation Loss")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Basic data preparation

with open('shakespeare.txt', 'r') as f:
    text = f.read()

# find the most common speakersin the text
from collections import Counter
import re
import numpy as np
speakers = []
for line in text.splitlines():
    match = re.match(r'^([A-Z\s]+):$', line)
    if match:
        speakers.append(match.group(1))

speaker_counts = Counter(speakers)
common_speakers = np.array(speaker_counts.most_common(10)).T[0]
```

```python

all_dialogues = text.split('\n\n')
```

```python
# create a dataset for speaker identification
speaker_identification_dataset = []
for speaker in common_speakers:
    for block in all_dialogues:
        if block.startswith(speaker + ':'):
            input = '[SPEAKER] ' + block[len(speaker) + 1:].strip() + ' [ANSWER] ' + speaker + ' [END]'
            speaker_identification_dataset.append(input)

print(len(speaker_identification_dataset))   
```

```python
# create a dataset for verse vs prose classification
verse_prose_dataset = []
for block in all_dialogues:
    if block.startswith('['):
        continue
    if re.search(r'[A-Z\s]+:', block):
        continue
    if re.search(r'\n', block):
        input = '[VERSE] ' + block.strip() + ' [ANSWER] VERSE [END]'
        verse_prose_dataset.append(input)
    else:
        input = '[PROSE] ' + block.strip() + ' [ANSWER] PROSE [END]'
        verse_prose_dataset.append(input)

for i in range(5):
    print(verse_prose_dataset[i])  
```
